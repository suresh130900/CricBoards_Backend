List of all the teams that are participating

All the Venue List

```python
all_venues = matches['venue'].tolist()
all_venues = list(set(all_venues))
all_venues
```

Number of Matches per Venue


```python
sns.countplot(x='venue', data=matches)
plt.xticks(rotation='vertical')
plt.show()
```

Matches palyed by Each Team


```python
x = matches['team1'].value_counts()
y = matches['team2'].value_counts()
x
y
```

Plotting all the Teams matches played

```python
(x+y).plot(kind='barh')
```

Matches Won by Each Team


```python
all_matches_won = pd.DataFrame({"Winner":matches['winning_team']}).value_counts()
all_matches_won
```

Plotting the graph of matches won

```python
sns.countplot(x = "winning_team", data=matches)
plt.xticks(rotation = 'vertical')
plt.show()
```

Top 5 Man of the Match

```python
MoM = matches['player_of_match'].value_counts().head()
MoM
```

Plotting MoM


```python
sns.barplot(x = MoM.index, y = MoM.values, data=matches)
plt.title("Top 5 MoM")
plt.xticks(rotation=90)
plt.xlabel("Match Count")
plt.ylabel("Player")
plt.show()
```

The Top Batsman in the IPL

```python
top_batsman=delivery.groupby('batter')['batsman_run'].agg('sum').reset_index().sort_values('batsman_run', ascending=False).head(10)
top_batsman.set_index('batter', inplace=True)
top_batsman.plot(kind='bar')
```

The Bowler Who Has Given the Highest Number of Runs

```python
worst_bowler = delivery.groupby('bowler')['total_run'].agg('sum').reset_index().sort_values('total_run', ascending=False).head(10)
worst_bowler.plot(kind='bar')
```

The Bowler with Team-wise Performance

```python
mask=delivery['bowler']=='PP Chawla'
delivery[mask].groupby('batting_team')['total_run'].agg('sum').plot(kind='bar')
```

Over-wise Batting Performance of Each Team in the IPL (2008–2020)

```python
delivery6=delivery[mask]
delivery6=delivery6[['batting_team','overs','batsman_run']]
x=delivery6.pivot_table(values='batsman_run', index='batting_team', columns='overs', aggfunc='count')
sns.heatmap(x, cmap='summer')
```

```python
sns.countplot(x = 'dismisal_kind', data=delivery)
plt.xticks(rotation='vertical')
```
